### Decision bias and memory

```python
pairs_acc = df_memory.groupby(["PID", "condition"]).agg(pairs_acc = ('pairs_acc', 'mean')).reset_index()
pairs_acc["condition"] = pairs_acc["condition"].replace({0: "Regular", 1: "Interference"})
p_gain_memory = p_gain.merge(pairs_acc, on=["PID","condition"], how="inner")
```

Assess inverse bias and rt (chosen minus unchosen pairs) as a function of memory

```python
p_gain_spread = p_gain_memory.drop(columns=["zscored_rt","pairs_acc"]).pivot_table(index=["PID","condition"],columns="chosen_trial",values="p_gain").reset_index()
p_gain_spread["inverse bias"] = p_gain_spread["Chosen Pairs"] - p_gain_spread["Unchosen Pairs"]
rt_spread = p_gain_memory.drop(columns=["p_gain","pairs_acc"]).pivot_table(index=["PID","condition"],columns="chosen_trial",values="zscored_rt").reset_index()
rt_spread["inverse zscored rt"] = rt_spread["Chosen Pairs"] - rt_spread["Unchosen Pairs"]
inverse_performance = p_gain_spread.drop(columns=["Chosen Pairs","Unchosen Pairs"]).merge(
    rt_spread.drop(columns=["Chosen Pairs","Unchosen Pairs"]), on=["PID","condition"]).merge(
    pairs_acc, on=["PID","condition"])
```

```python
sns.lmplot(data=p_gain_memory, x="pairs_acc", y="p_gain", col="chosen_trial", hue="condition", ci=95, truncate=False);

sns.lmplot(data=inverse_performance, x="pairs_acc", y="inverse bias", hue="condition", ci=95, truncate=False, aspect=1.2).set(xlabel="Pairs memory (accuracy)", ylabel="Inverse bias\n(p(gain) chosen - p(gain) unchosen)");

sns.lmplot(data=inverse_performance, x="pairs_acc", y="inverse zscored rt", hue="condition", ci=95, truncate=False, aspect=1.2).set(xlabel="Pairs memory (accuracy)", ylabel="Inverse zscored RT\n(RT chosen - RT unchosen)");

```

Now assess individual differences in memory and p(gain) between the different conditions 

```python
cond_spread_p_gain = p_gain_memory.drop(columns=["zscored_rt","pairs_acc"]).pivot_table(index=["PID","chosen_trial"],columns="condition",values="p_gain").reset_index()
cond_spread_p_gain["condition diff - choices"] = cond_spread_p_gain["Regular"] - cond_spread_p_gain["Interference"]
cond_spread_rt = p_gain_memory.drop(columns=["p_gain","pairs_acc"]).pivot_table(index=["PID","chosen_trial"],columns="condition",values="zscored_rt").reset_index()
cond_spread_rt["condition diff - zscored rt"] = cond_spread_rt["Regular"] - cond_spread_rt["Interference"]
cond_spread_pairs_acc = pairs_acc.pivot_table(index=["PID"],columns="condition",values="pairs_acc").reset_index()
cond_spread_pairs_acc["condition diff - pairs_acc"] = cond_spread_pairs_acc["Regular"] - cond_spread_pairs_acc["Interference"]

cond_diff_performance = cond_spread_p_gain.drop(columns=["Regular","Interference"]).merge(
    cond_spread_rt.drop(columns=["Regular","Interference"]), on=["PID","chosen_trial"]).merge(
    cond_spread_pairs_acc.drop(columns=["Regular","Interference"]), on=["PID"])
```

```python
sns.lmplot(data=cond_diff_performance, x="condition diff - pairs_acc", y="condition diff - choices", hue="chosen_trial", ci=95, truncate=False, aspect=1.5).set(xlabel="Pairs memory difference\n(Regular - Interference)", ylabel="p(gain) difference\n(Regular - Interference)");
sns.lmplot(data=cond_diff_performance, x="condition diff - pairs_acc", y="condition diff - zscored rt", hue="chosen_trial", ci=95, truncate=False, aspect=1.5).set(xlabel="Pairs memory difference\n(Regular - Interference)", ylabel="zscored RT difference\n(Regular - Interference)");
```

## Choice probability as a function of kind type
Now we can look at the same result, but with binning delta value into groups of 20. This will allow us to look at average performance across participants.

```python
# function to plot choice and rt as a function of delta value
def plot_catplot(df, x_var, y_var, hue_var, col_var, line, title, y_label, x_label, ci_var):
    
    # no col grouping
    if not col_var:
        group_binned_data = df.groupby(["PID",hue_var,x_var]).agg(y_group = (y_var, "mean")).reset_index()
        g = sns.catplot(data = group_binned_data, x=x_var, y="y_group", hue=hue_var, dodge=True, kind="point", ci=ci_var)
    # add col grouping
    else:
        group_binned_data = df.groupby(["PID",hue_var,col_var,x_var]).agg(y_group = (y_var, "mean")).reset_index()
        g = sns.catplot(data = group_binned_data, x=x_var, y="y_group", hue=hue_var, col=col_var, dodge=True, kind="point", ci=ci_var)
    
    if line:
        g.map(plt.axhline, y=line, ls='--', c='black')
        
    # add axes labels and title
    g.fig.subplots_adjust(top=0.8);
    g.fig.suptitle(title);
    g.set_axis_labels(x_label,y_label);

    return g;
```

```python
# plot choice probability 
plot_catplot(df_decisions, "binned_delta_comulative_val_cat","left_chosen", "kind_type", [], 0.5, "Choice probability as a function of kind type", "p(Choose Left)","\u0394 comulative value", 95)

# across blocks
plot_catplot(df_decisions, "binned_delta_comulative_val_cat","left_chosen", "kind_type", "block", 0.5, "Choice probability across blocks", "p(Choose Left)","\u0394 comulative value", 95)

# plot choice probability as a function of delta grouping value
plot_catplot(df_decisions, "delta_group_val_cat","left_chosen", "kind_type", [], 0.5, "Choice probability as a function of kind type", "p(Choose Left)","\u0394 grouping value", 95)

# across blocks
plot_catplot(df_decisions, "delta_group_val_cat","left_chosen", "kind_type", "block", 0.5, "Choice probability across blocks", "p(Choose Left)","\u0394 grouping value", 95)
```

## Probability of choosing a category as a function of its value


```python
# we compute mean rt and choice probability for the different categories. 
# we do this by first separating the data into chosen and unchosen categories, and then combine them together.

df_chosen = df_decisions[["PID", "block", "kind_type","chosen_cat", "rt", "zscored_rt", "chosen_cat_val_group","chosen_exemplar_val"]].assign(
    category = df_decisions.chosen_cat,
    cat_value_group = df_decisions.chosen_cat_val_group,
    chosen = 1).drop(
    columns = ["chosen_cat", "chosen_cat_val_group"])
# for unchosen categories we remove value information because it was not given, and the rt column so it won't be duplicated
df_unchosen = df_decisions[["PID", "block", "kind_type", "unchosen_cat_val_group","unchosen_cat","zscored_rt", "rt"]].assign(
    category = df_decisions.unchosen_cat,
    cat_value_group = df_decisions.unchosen_cat_val_group,
    chosen = 0, 
    chosen_exemplar_val=np.nan,
    rt = np.nan,
    zscored_rt = np.nan).drop(
    columns = ["unchosen_cat", "unchosen_cat_val_group"])
df_combined = df_chosen.append(df_unchosen).reset_index()


# compute average value for each category
category_values = df_combined.groupby(['PID','category','kind_type','cat_value_group']).agg(
    comulative_cat_value = ('chosen_exemplar_val','mean')).reset_index()

# compute rt and choice prbability per condition 
category_performance = df_combined.groupby(['PID','kind_type','block','cat_value_group','category']).agg(
    mean_rt = ('rt' , 'mean'),
    mean_zscored_rt = ('zscored_rt', 'mean'),
    choice_prob = ('chosen', 'mean')).reset_index()

# merge these two data sets 
categories = category_performance.merge(category_values, on=['PID','category','kind_type','cat_value_group'])
```

```python
sns.catplot(data = categories, 
            x="cat_value_group", y="choice_prob", hue='kind_type', kind="point", err_style="bars", ci=95).map(
    plt.axhline, y=0.5, ls='--', c='black').set(
    ylabel="p(Choose Category)",xlabel="Category Value", title="Category effect as a function of kind type")
```

## Prediction error analysis

Here we want to test whether getting unexpected feedback results in any behavioral consequences. I.e., if I expect the bowtie category to result in one dollar value, and I get 60 cents, will that affect my behavior. 
Here are some ideas on the behavioral consequences:
1. Surpising feedback may slow down the next response 
2. negative prediction error may result in avoidance to choose an item from that category on the next trial it appears
3. positive prediction error may result in approach to choosing an item from the category on the next trial

```python
# compute prediction error (difference between shown exemplar value and the learned comulative category value)
df_decisions["cat_prediction_error"] = df_decisions["chosen_exemplar_val"] - df_decisions["cat_comulative_val_chosen"]

# check whether a category was chosen on trial n+1 if it was chosen on trial n
for index, row in df_decisions.iloc[0:-1].iterrows():
    # find all trials where the chosen category on t=n was presented
    curr_cat_trials = df_decisions.loc[(df_decisions.PID == df_decisions.loc[index, "PID"]) & ((df_decisions.chosen_cat == df_decisions.loc[index, "chosen_cat"]) | (df_decisions.unchosen_cat == df_decisions.loc[index, "chosen_cat"]))]
    # out of these trials, find trial n+1 
    curr_exemplar = curr_cat_trials.loc[index,"chosen_exemplar"]
    curr_cat_trials = curr_cat_trials.reset_index()
    if (curr_cat_trials.index[curr_cat_trials.chosen_exemplar == curr_exemplar] + 1) < len(curr_cat_trials):
        trial_n_plus_one = curr_cat_trials.iloc[curr_cat_trials.index[curr_cat_trials.chosen_exemplar == curr_exemplar] + 1]
    else:
        trial_n_plus_one = pd.DataFrame();

    # if the same category was chosen on trial t=n+1 then add 1
    if not trial_n_plus_one.empty:
        if (trial_n_plus_one["chosen_cat"] == df_decisions.loc[index, "chosen_cat"]).bool():
            df_decisions.loc[index, "cat_chosen_n_plus_one"] = 1;
        else:
            df_decisions.loc[index, "cat_chosen_n_plus_one"] = 0;
    else:
        df_decisions.loc[index, "cat_chosen_n_plus_one"] = np.nan;
```

```python
# bin prediction error
df_decisions = df_decisions.assign(
    binned_cat_prediction_error = pd.cut(df_decisions.cat_prediction_error,
                                              bins=[-40, -25, -15, -5, 5, 15, 25, 40],
                                              labels=["-30","-15","-7.5","0","7.5","15","30"]))
```

## Test good/bad hypothesis
According to this hypothesis, people do not maintain a comulative value of all categories. Instead, they classify the categories to good and bad if they encounter a very good exemplar (100 cent) or a very bad one (0 ceny). Those in the middle do not get any classification (or they are the middle class). To test this hypothesis, for the right and left items we incoude a column asking what tyoe of category is it - good (1), bad (-1) or don't know (0). Weh then compute the difference between the two items and assess whether the outcome predicts choices. 

```python
# here we categorize the left and the right exemplars as belonging to a good or a bad category 
# according to all previous presentations of their category
good_exemplar_val = 100;
bad_exemplar_val = 0;
for index, row in df_decisions.iterrows():
    # find all trial before the current trials where this category was chosen
    left_cat_trials = df_decisions.loc[(df_decisions.PID == df_decisions.loc[index, "PID"]) & (df_decisions.chosen_cat == df_decisions.loc[index, "cat_left"])].loc[:(index-1)]
    right_cat_trials = df_decisions.loc[(df_decisions.PID == df_decisions.loc[index, "PID"]) & (df_decisions.chosen_cat == df_decisions.loc[index, "cat_right"])].loc[:(index-1)]
    
    # check if left_cat_trials is empty
    if left_cat_trials.empty:
        df_decisions.loc[index, "left_good_cat"] = np.nan;
        df_decisions.loc[index, "left_bad_cat"] = np.nan;
    else:
        if any(bad_exemplar_val >= np.array(left_cat_trials.chosen_exemplar_val)):
            df_decisions.loc[index, "left_bad_cat"] = 1;
            df_decisions.loc[index, "left_good_cat"] = 0;
        elif any(good_exemplar_val <= np.array(left_cat_trials.chosen_exemplar_val)):
            df_decisions.loc[index, "left_good_cat"] = 1;
            df_decisions.loc[index, "left_bad_cat"] = 0;
        else:
            df_decisions.loc[index, "left_good_cat"] = 0;
            df_decisions.loc[index, "left_bad_cat"] = 0;
            
    # check if left_cat_trials is empty
    if right_cat_trials.empty:
        df_decisions.loc[index, "right_good_cat"] = np.nan;
        df_decisions.loc[index, "right_bad_cat"] = np.nan;
    else:
        if any(bad_exemplar_val >= np.array(right_cat_trials.chosen_exemplar_val)):
            df_decisions.loc[index, "right_bad_cat"] = 1;
            df_decisions.loc[index, "right_good_cat"] = 0;
        elif any(good_exemplar_val <= np.array(right_cat_trials.chosen_exemplar_val)):
            df_decisions.loc[index, "right_good_cat"] = 1;
            df_decisions.loc[index, "right_bad_cat"] = 0;
        else:
            df_decisions.loc[index, "right_good_cat"] = 0;
            df_decisions.loc[index, "right_bad_cat"] = 0;
```

```python
# add a response column where choices are made using the bad and good cats columns
# choose the left/right if the left/right include a good category (approach) or the other one includes a bad category (avoid)
# if neither is true (say, the categories are middle ones or its a new category with no prior value history) choose randomly
for index, row in df_decisions.iterrows():
    if (df_decisions.loc[index, "left_good_cat"]==1) | (df_decisions.loc[index, "right_bad_cat"]==1):
        df_decisions.loc[index, "predicted_left_chosen"] = 1;
        df_decisions.loc[index, "good_bad_cat_info"] = 1;
    elif (df_decisions.loc[index, "right_good_cat"]==1) | (df_decisions.loc[index, "left_bad_cat"]==1):
        df_decisions.loc[index, "predicted_left_chosen"] = 0;
        df_decisions.loc[index, "good_bad_cat_info"] = 1;
    else:
        df_decisions.loc[index, "good_bad_cat_info"] = 0;
        rand_number = random.uniform(0, 1)
        if rand_number > 0.5:
            df_decisions.loc[index, "predicted_left_chosen"] = 1;
        else:
            df_decisions.loc[index, "predicted_left_chosen"] = 0;
            
```

```python
# test the classification accuracy of this model
def f(x): 
    return np.nanmean(x.predicted_left_chosen == x.left_chosen)
p_class = df_decisions.groupby(["PID","good_bad_cat_info"]).apply(f).reset_index()
p_class.columns = ["PID", "good_bad_cat_info","class_acc"]
sns.catplot(data = p_class, x="good_bad_cat_info", y="class_acc", dodge=True, kind="point", aspect=1.5)

p_class2 = df_decisions.groupby(["PID"]).apply(f).reset_index()
p_class2.columns = ["PID", "class_acc"]
sns.histplot(data=p_class2, x="class_acc").set(xlabel="Classification accuracy", ylabel="Count")

plt.axvline(0.5, color='r', linestyle="--");
```

## Testing whether the learning the high category values is enough
Here we want to look at the effect of category learning as a function of the category values. That is, we want to test whether it is enough to only learn the values of the high and/or low category values to be able to perform well in the experiment 


```python
# categorize trials by the category value groups 
for index, row in df_decisions.iterrows():
    if (df_decisions.loc[index,"cat_val_right"] == 80) | (df_decisions.loc[index,"cat_val_left"] == 80):
        df_decisions.loc[index,"value_group_trial"] = "high";
    elif (df_decisions.loc[index,"cat_val_right"] == 20) | (df_decisions.loc[index,"cat_val_left"] == 20):
        df_decisions.loc[index,"value_group_trial"] = "low";
    else: 
        df_decisions.loc[index,"value_group_trial"] = "middle";
```

In this phase we assess how well participants remembered the actual values shown as well as if the item was old or new and whether they chose it. 
We then try to see whether their memory predicts their tendency to learn and use category values.

```python
# create a table with mean acc and rt for the different memory tests
exemplar_memory_acc = pd.melt(df_exemplar_memory, id_vars=["PID","kind_type","cat_value_group"],value_vars=["is_old_acc", "is_chosen_acc", "exemplar_value_acc"], var_name='exemplar_memory_type', value_name='acc')
exemplar_memory_acc['exemplar_memory_type'] = exemplar_memory_acc['exemplar_memory_type'].str.replace("is_old_acc", "old_memory").str.replace("is_chosen_acc", "choice_memory").str.replace("exemplar_value_acc", "value_memory")
exemplar_memory_rt = pd.melt(df_exemplar_memory, id_vars=["PID","kind_type","cat_value_group"],value_vars=["zscored_rt_old_memory", "zscored_rt_choice_memory", "zscored_rt_value_memory"], var_name='exemplar_memory_type', value_name='zscored_rt')
exemplar_memory_rt['exemplar_memory_type'] = exemplar_memory_rt['exemplar_memory_type'].str.replace("zscored_rt_", "")
exemplar_memory_all = exemplar_memory_acc.merge(exemplar_memory_rt, on=["PID","kind_type","cat_value_group","exemplar_memory_type"])
exemplar_memory_grouping = exemplar_memory_all.groupby(["PID","kind_type","cat_value_group","exemplar_memory_type"]).agg(acc = ('acc','mean'), zscored_rt = ('zscored_rt','mean')).reset_index()
```
