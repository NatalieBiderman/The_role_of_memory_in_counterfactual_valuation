# Preprocess data and clean outliers
We use a function to preprocess the data (loading individual data, concatenating them, and seperating into the different experimental phases). 
We start by loading data from all participants, then we check for outliers and remove them. 

```python
Exp = "Exp1"
version="Version11"
process_mturk_csv('../../../../../Art_experiment/Java_experiment/' + version + '/Data', 'condition', Exp)
```

```python
def preprocess_data_by_exp(Exp):
    
    if (Exp == "Exp1"):
        version = "Version11"
    else: 
        version = "Version10"
        
    # preprocess
    process_mturk_csv('../../../../../Art_experiment/Java_experiment/' + version + '/Data', 'condition', Exp)
    
    # load preprocessed csvs
    keys = ["ratings", "deliberation", "interference", "reward_learning", "final_decisions", "memory", "outcome_evaluation", "all_data", "all_interaction_data", "debrief"]
    path_to_data = "../../Data/" + Exp + "/Summary_data/all_subs"
    dfs_all_subs = {} # create a dictionary

    for i, key in enumerate(keys):
        if os.path.isfile(os.path.join(path_to_data, key + ".csv")):
            dfs_all_subs[key] = pd.read_csv(os.path.join(path_to_data, key + ".csv"))
            if "Unnamed: 0" in dfs_all_subs[key].columns:
                dfs_all_subs[key] = dfs_all_subs[key].drop(columns = "Unnamed: 0")
            
    # find outlier subjects
    blur_focus_criterion = 25;
    missed_instructions_criterion = 10; 
    final_decisions_non_responses_criterion = 25; 
    full_screen_criterion = 10;
    below_chance_chosen_final_decisions_criterion = 0.5;
    warning_events, outlier_subs = find_outlier_subs(dfs_all_subs["final_decisions"], dfs_all_subs["all_data"], dfs_all_subs["all_interaction_data"], blur_focus_criterion, missed_instructions_criterion, final_decisions_non_responses_criterion, full_screen_criterion, below_chance_chosen_final_decisions_criterion)
    print("outlier subs:")
    print(outlier_subs)
    
    # remove outlier subjects
    keys = ["ratings", "deliberation", "interference", "reward_learning", "final_decisions", "memory", "outcome_evaluation", "all_data", "all_interaction_data", "debrief"]
    dfs = {} # create a dictionary
    for i, key in enumerate(keys):
        if key in dfs_all_subs:
            dfs[key] = remove_outlier_subs(outlier_subs, dfs_all_subs[key], key, Exp)
            
    # remove non responses from decisions df
    dfs["final_decisions"] = dfs["final_decisions"][~pd.isnull(dfs["final_decisions"]["rt"])]
    # show how many subjects were left 
    print("after removal of outliers, n = " + str(len(np.unique(dfs["final_decisions"].PID))))
```

```python
preprocess_data_by_exp("Pilot")
preprocess_data_by_exp("Exp1")
```

# Load clean data

```python
keys = ["ratings", "deliberation", "interference", "reward_learning", \
        "final_decisions", "memory", "outcome_evaluation", "all_data", "all_interaction_data", "debrief"]
path_to_data = "../Data/Exp1/Summary_data/non_outlier_subs"
dfs = {} # create a dictionary

for i, key in enumerate(keys):
    dfs[key] = pd.read_csv(os.path.join(path_to_data, key + ".csv"))
    if "Unnamed: 0" in dfs[key].columns:
            dfs[key] = dfs[key].drop(columns = "Unnamed: 0")

# remove non responses from final decisions
dfs["final_decisions"] = dfs["final_decisions"][~pd.isnull(dfs["final_decisions"]["rt"])]
```

## Analysis of decision bias
first let's look at inverse bias across the conditions and then within them. We were not able to replicate the inverse bias in neither. We do see an interesting RT effect for chosen pairs - responses to Repetition pairs are faster than Interference pairs, suggesting some enhanced learning of reward for chosen pairs. This is the only difference we found between conditions.

### Decision bias and memory
This was the main effect replicated across different tasks: a strong correlation between pairs memory and inverse bias (chosen minus unchosen pairs). Here this effect seems to replicate, but without a difference between conditions 

```python
p_gain_memory = p_gain.merge(pairs_acc.drop(columns="zscored_rt"), on=["PID","condition"], how="inner")
#p_gain_memory
```

```python
p_gain_spread = p_gain_memory.drop(columns=["pairs_acc", "zscored_rt"]).pivot_table(index=["PID","condition"],columns="chosen_trial",values="p_gain").reset_index()
p_gain_spread["inverse bias"] = p_gain_spread["Chosen Pairs"] - p_gain_spread["Unchosen Pairs"]
rt_spread = p_gain_memory.drop(columns=["p_gain","pairs_acc"]).pivot_table(index=["PID","condition"],columns="chosen_trial",values="zscored_rt").reset_index()
rt_spread["inverse zscored rt"] = rt_spread["Chosen Pairs"] - rt_spread["Unchosen Pairs"]
inverse_performance = p_gain_spread.drop(columns=["Chosen Pairs","Unchosen Pairs"]).merge(
   rt_spread.drop(columns=["Chosen Pairs","Unchosen Pairs"]), on=["PID","condition"]).merge(
   pairs_acc, on=["PID","condition"])
```

```python
ax = sns.catplot(data = inverse_performance, 
                x="condition", y="inverse bias", kind="bar", ci=95, alpha=0.5, capsize=.2).map(
        plt.axhline, y=0.5, ls='--', c='black').set(
        ylabel="Inverse bias\n(p(gain) chosen - p(gain) unchosen)",xlabel="Pair type")
ax = sns.swarmplot(data=inverse_performance, x="condition", y="inverse bias", dodge=True)
#ax.get_legend().set_visible(False)
```

```python
sns.lmplot(data=p_gain_memory, x="pairs_acc", y="p_gain", col="chosen_trial", hue="condition", ci=95, truncate=False);

sns.lmplot(data=inverse_performance, x="pairs_acc", y="inverse bias", hue="condition", ci=95, truncate=False, aspect=1.2).set(xlabel="Pairs memory (accuracy)", ylabel="Inverse bias\n(p(gain) chosen - p(gain) unchosen)");
plt.savefig('Plots/memory_bias.png', dpi=300)

sns.lmplot(data=inverse_performance, x="pairs_acc", y="inverse zscored rt", hue="condition", ci=95, truncate=False, aspect=1.2).set(xlabel="Pairs memory (accuracy)", ylabel="Inverse zscored RT\n(RT chosen - RT unchosen)");

```

Now assess individual differences in memory and p(gain) between the different conditions 

```python
cond_spread_p_gain = p_gain_memory.drop(columns=["zscored_rt","pairs_acc"]).pivot_table(index=["PID","chosen_trial"],columns="condition",values="p_gain").reset_index()
cond_spread_p_gain["condition diff - choices"] = cond_spread_p_gain["Repetition"] - cond_spread_p_gain["Interference"]
cond_spread_rt = p_gain_memory.drop(columns=["p_gain","pairs_acc"]).pivot_table(index=["PID","chosen_trial"],columns="condition",values="zscored_rt").reset_index()
cond_spread_rt["condition diff - zscored rt"] = cond_spread_rt["Repetition"] - cond_spread_rt["Interference"]
cond_spread_pairs_acc = pairs_acc.pivot_table(index=["PID"],columns="condition",values="pairs_acc").reset_index()
cond_spread_pairs_acc["condition diff - pairs_acc"] = cond_spread_pairs_acc["Repetition"] - cond_spread_pairs_acc["Interference"]

cond_diff_performance = cond_spread_p_gain.drop(columns=["Repetition","Interference"]).merge(
    cond_spread_rt.drop(columns=["Repetition","Interference"]), on=["PID","chosen_trial"]).merge(
    cond_spread_pairs_acc.drop(columns=["Repetition","Interference"]), on=["PID"])
```

# Divide subjects to two groups by performance

```python
# divide subjects into two groups according to their performance in chosen pairs 
p_gain_chosen = dfs["final_decisions"].loc[dfs["final_decisions"].chosen_trial==1].groupby(["PID"]).agg(
    p_gain = ('higher_outcome_chosen','mean')).sort_values(by="p_gain").reset_index()
p_gain_chosen.loc[0:np.floor(len(p_gain_chosen)/2), "performance_group"] = "low"
p_gain_chosen.loc[np.ceil(len(p_gain_chosen)/2):len(p_gain_chosen), "performance_group"] = "high"
performance_group = p_gain_chosen[["PID","performance_group"]]

dfs["final_decisions"] = dfs["final_decisions"].merge(performance_group, on="PID")
dfs["memory"] = dfs["memory"].merge(performance_group, on="PID")
dfs["outcome_evaluation"] = dfs["outcome_evaluation"].merge(performance_group, on="PID")

# here we decide whether we want to remove the low performance subjects from further analysis
apply_strict_exclusion_criteria = 0;
if (apply_strict_exclusion_criteria==1):
    # find low performance subjects 
    chosen_p_gain_cutoff = 0.5
    memory_acc_cutoff = 0.5
    p_gain_chosen = dfs["final_decisions"].groupby(["PID","chosen_trial"]).agg(
        p_gain = ('higher_outcome_chosen','mean')).reset_index()
    low_performance_subs_p_gain = list(p_gain_chosen.loc[(p_gain_chosen.chosen_trial==1) & (p_gain_chosen.p_gain < chosen_p_gain_cutoff), "PID"])
    pairs_acc_high = dfs["memory"].groupby(["PID"]).agg(
        pairs_acc = ('pairs_acc','mean')).reset_index()
    low_performance_subs_pairs_acc = list(pairs_acc_high.loc[pairs_acc_high.pairs_acc < memory_acc_cutoff, "PID"])
    low_performance_subs = np.unique(low_performance_subs_p_gain + low_performance_subs_pairs_acc)
    # remove these subjects from dfs
    dfs["final_decisions"] = dfs["final_decisions"].loc[~dfs["final_decisions"].PID.isin(low_performance_subs)]
    dfs["memory"] = dfs["memory"].loc[~dfs["memory"].PID.isin(low_performance_subs)]
    dfs["outcome_evaluation"] = dfs["outcome_evaluation"].loc[~dfs["outcome_evaluation"].PID.isin(low_performance_subs)]
```

```python
# memory

pairs_acc = dfs["memory"].groupby(["PID", "performance_group","condition"]).agg(pairs_acc = ('pairs_acc', 'mean'),
                                                                                zscored_rt = ('zscored_rt_pairs', 'mean')).reset_index()
pairs_acc["condition"] = pairs_acc["condition"].replace({0: "Repetition", 1: "Interference"})

pairs_acc.groupby(["performance_group","condition",]).agg({'pairs_acc': ('mean', 'sem')}).reset_index()

ax = sns.catplot(data = pairs_acc, 
                x="condition", y="pairs_acc", kind="bar", ci=95, alpha=0.5, capsize=.2, col="performance_group").map(
        plt.axhline, y=0.5, ls='--', c='black').set(
        ylabel="Memory accuracy",xlabel="Condition")
bx = sns.catplot(data = pairs_acc, 
                x="condition", y="zscored_rt", kind="bar", ci=95, alpha=0.5, capsize=.2, col="performance_group").map(
        plt.axhline, y=0, ls='--', c='black').set(
        ylabel="RT (zscored)",xlabel="Condition")

# choices in final decisions
p_gain = dfs["final_decisions"].groupby(["PID","performance_group","chosen_trial","condition"]).agg(
    p_gain = ('higher_outcome_chosen','mean'),
    zscored_rt = ('zscored_rt', 'mean')).reset_index()
p_gain["chosen_trial"] = p_gain["chosen_trial"].replace({0: "Unchosen Pairs", 1: "Chosen Pairs"})
p_gain["condition"] = p_gain["condition"].replace({0: "Repetition", 1: "Interference"})
ax = sns.catplot(data = p_gain, 
                x="chosen_trial", y="p_gain", col="performance_group",kind="bar", ci=95, aspect=1.5, alpha=0.5, capsize=.2).map(
        plt.axhline, y=0.5, ls='--', c='black').set(
        ylabel="p(Choose Gain)",xlabel="Pair type")
#ax = sns.swarmplot(data=p_gain, x="chosen_trial", y="p_gain", col="performance_group", dodge=True)
#ax.get_legend().set_visible(False)
sns.catplot(data = p_gain, 
                x="chosen_trial", y="p_gain", hue="condition", kind="bar", col="performance_group",ci=95, aspect=1.5, alpha=0.5, capsize=.2).map(
        plt.axhline, y=0.5, ls='--', c='black').set(
        ylabel="p(Choose Gain)",xlabel="Pair type")

# rt in final decisions
rt_gain_choice = dfs["final_decisions"].groupby(["PID","performance_group","chosen_trial","condition","higher_outcome_chosen"]).agg(
    zscored_rt = ('zscored_rt', 'mean')).reset_index()
rt_gain_choice["chosen_trial"] = rt_gain_choice["chosen_trial"].replace({0: "Unchosen Pairs", 1: "Chosen Pairs"})
rt_gain_choice["condition"] = rt_gain_choice["condition"].replace({0: "Repetition", 1: "Interference"})
ax = sns.catplot(data = rt_gain_choice.loc[rt_gain_choice.chosen_trial=="Unchosen Pairs"], 
                col="performance_group", hue="higher_outcome_chosen", y="zscored_rt", x='condition', kind="bar", ci=95, aspect=1.5, alpha=0.5, capsize=0.1).map(
        plt.axhline, y=0, ls='--', c='black').set(
        ylabel="z-scored RT",xlabel="Gain chosen")

# inverse bias and memory correlation
p_gain_memory = p_gain.merge(pairs_acc.drop(columns="zscored_rt"), on=["PID","performance_group","condition"], how="inner")

p_gain_spread = p_gain_memory.drop(columns=["pairs_acc", "zscored_rt"]).pivot_table(index=["PID","condition","performance_group"],columns="chosen_trial",values="p_gain").reset_index()
p_gain_spread["inverse bias"] = p_gain_spread["Chosen Pairs"] - p_gain_spread["Unchosen Pairs"]
rt_spread = p_gain_memory.drop(columns=["p_gain","pairs_acc"]).pivot_table(index=["PID","condition","performance_group"],columns="chosen_trial",values="zscored_rt").reset_index()
rt_spread["inverse zscored rt"] = rt_spread["Chosen Pairs"] - rt_spread["Unchosen Pairs"]
inverse_performance = p_gain_spread.drop(columns=["Chosen Pairs","Unchosen Pairs"]).merge(
   rt_spread.drop(columns=["Chosen Pairs","Unchosen Pairs"]), on=["PID","condition","performance_group"]).merge(
   pairs_acc, on=["PID","condition","performance_group"])

sns.lmplot(data=p_gain_memory.loc[p_gain_memory.chosen_trial=="Unchosen Pairs"], x="pairs_acc", y="p_gain", col="performance_group", hue="condition", ci=95, truncate=False).set(ylabel="p(select S+)\nUnchosen pairs");
sns.lmplot(data=p_gain_memory.loc[p_gain_memory.chosen_trial=="Chosen Pairs"], x="pairs_acc", y="p_gain", col="performance_group", hue="condition", ci=95, truncate=False).set(ylabel="p(select S+)\nChosen pairs");

sns.lmplot(data=inverse_performance, x="pairs_acc", y="inverse bias", hue="condition", col="performance_group",ci=95, truncate=False, aspect=1).set(xlabel="Pairs memory (accuracy)", ylabel="Inverse bias\n(p(gain) chosen - p(gain) unchosen)");


# outcome evaluation
outcome_eval_summ = dfs["outcome_evaluation"].groupby(["PID", "performance_group","stim_type","condition", "outcome"]).agg(
    gain_estimation = ("outcome_eval_gain", "mean"),
    gain_estimation_rt = ("zscored_outcome_eval_rt", "mean"),
    confidence = ("outcome_eval_confidence","mean"),
    confidence_rt = ("zscored_outcome_eval_confidence_rt","mean")).reset_index()
outcome_eval_summ["condition"] =  outcome_eval_summ["condition"].replace({0: "Repetition", 1: "Interference"})
outcome_eval_summ["outcome"] =  outcome_eval_summ["outcome"].replace({-1: "No gain", 1: "Gain"})

ax = sns.catplot(data = outcome_eval_summ.loc[outcome_eval_summ.performance_group=="high"], 
                x="outcome", y="gain_estimation", hue="condition", col="stim_type", kind="bar", ci=95, aspect=1.5, alpha=0.5, capsize=0.1).map(
        plt.axhline, y=0.5, ls='--', c='black').set(
        ylabel="p(estimate as S+)\nHigh group")

ax = sns.catplot(data = outcome_eval_summ.loc[outcome_eval_summ.performance_group=="low"], 
                x="outcome", y="gain_estimation", hue="condition", col="stim_type",kind="bar", ci=95, aspect=1.5, alpha=0.5, capsize=0.1).map(
        plt.axhline, y=0.5, ls='--', c='black').set(
        ylabel="p(estimate as S+)\nLow group")

ax = sns.catplot(data = outcome_eval_summ.loc[outcome_eval_summ.performance_group=="high"], 
                x="outcome", y="gain_estimation_rt", hue="condition", col="stim_type", kind="bar", ci=95, aspect=1.5, alpha=0.5, capsize=0.1).map(
        plt.axhline, y=0, ls='--', c='black').set(
        ylabel="RT (zscored)\nHigh group")

ax = sns.catplot(data = outcome_eval_summ.loc[outcome_eval_summ.performance_group=="low"], 
                x="outcome", y="gain_estimation_rt", hue="condition", col="stim_type", kind="bar", ci=95, aspect=1.5, alpha=0.5, capsize=0.1).map(
        plt.axhline, y=0, ls='--', c='black').set(
        ylabel="RT (zscored)\nLow group")

ax = sns.catplot(data = outcome_eval_summ.loc[outcome_eval_summ.performance_group=="high"], 
                x="outcome", y="confidence", hue="condition", col="stim_type", kind="bar", ci=95, aspect=1.5, alpha=0.5, capsize=0.1).map(
        plt.axhline, y=3.5, ls='--', c='black').set(
        ylabel="Confidence (1-6)\nHigh group")

ax = sns.catplot(data = outcome_eval_summ.loc[outcome_eval_summ.performance_group=="low"], 
                x="outcome", y="confidence", hue="condition", col="stim_type", kind="bar", ci=95, aspect=1.5, alpha=0.5, capsize=0.1).map(
        plt.axhline, y=3.5, ls='--', c='black').set(
        ylabel="Confidence (1-6)\nLow group")


```
